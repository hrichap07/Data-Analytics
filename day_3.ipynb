Calculate Mean, Median, Mode, Variance, Standard Deviation

```python
# Question 1
import statistics

numbers = [10, 20, 20, 30, 40, 50]

mean_value = statistics.mean(numbers)
median_value = statistics.median(numbers)
mode_value = statistics.mode(numbers)
variance_value = statistics.variance(numbers)
std_deviation = statistics.stdev(numbers)

print("Mean:", mean_value)
print("Median:", median_value)
print("Mode:", mode_value)
print("Variance:", variance_value)
print("Standard Deviation:", std_deviation)
```

2. Create a NumPy array and perform the following operations:

```python
# Question 2
import numpy as np

arr = np.array([1, 2, 3, 4, 5, 6])

print("Array:", arr)

# Sum and Mean
print("Sum:", np.sum(arr))
print("Mean:", np.mean(arr))

# Max and Min
print("Maximum:", np.max(arr))
print("Minimum:", np.min(arr))

# Reshape
reshaped_arr = arr.reshape(2,3)
print("Reshaped Array:\n", reshaped_arr)

# Element-wise multiplication
result = arr * 2
print("Element-wise multiplication:", result)
```

Write a Python program to:
•	Create two matrices
•	Perform addition and subtraction
•	Perform matrix multiplication
•	Find transpose of a matrix


```python
# Question 3
import numpy as np

matrix1 = np.array([[1,2,3],
                    [4,5,6]])

matrix2 = np.array([[7,8,9],
                    [1,2,3]])

print("Matrix 1:\n", matrix1)
print("Matrix 2:\n", matrix2)

# Addition
print("Addition:\n", matrix1 + matrix2)

# Subtraction
print("Subtraction:\n", matrix1 - matrix2)

# Matrix Multiplication
matrix3 = np.array([[1,2],
                    [3,4],
                    [5,6]])

multiplication = np.dot(matrix1, matrix3)
print("Matrix Multiplication:\n", multiplication)

# Transpose
print("Transpose of Matrix1:\n", matrix1.T)
```

Write a Python program using SciPy to:
•	Calculate mean using scipy.stats
•	Perform a t-test on a sample dataset


```python
# Question 5
import numpy as np
from scipy import stats

data = np.array([10, 12, 9, 11, 10, 13, 12])

# Mean using scipy
mean_value = stats.tmean(data)
print("Mean using scipy:", mean_value)

# One sample t-test
t_statistic, p_value = stats.ttest_1samp(data, 10)

print("T-Statistic:", t_statistic)
print("P-Value:", p_value)
```

. Write a Python program to compute definite integration of a mathematical function using scipy.integrate.

```python
# Question 6
from scipy import integrate
import numpy as np

# Function definition
def func(x):
    return x**2

# Definite integration from 0 to 3
result, error = integrate.quad(func, 0, 3)

print("Integration Result:", result)
print("Estimated Error:", error)
```

Write a Python program to minimize a mathematical function using scipy.optimize.



```python
# Question 7
from scipy.optimize import minimize

# Function to minimize
def func(x):
    return x**2 + 4*x + 4

initial_guess = 0

result = minimize(func, initial_guess)

print("Minimum Value:", result.fun)
print("Minimum at x =", result.x)
```
